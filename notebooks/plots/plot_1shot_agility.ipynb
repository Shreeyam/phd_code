```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(-30, 30, 100) # degrees
h = 400
v = 8 # km/s

rho = 1 # deg/sec

theta = np.radians(theta)
t = np.linspace(0, 10, 100)

# Targets
x = np.array([[8 * v, 0], [7.5 * v, np.deg2rad(20)], [7 * v, np.deg2rad(-20)]])
x_average = np.mean(x, axis=0)

X_t, Y_t = np.meshgrid(t, theta)

X = X_t * v
Y = Y_t * h

flat_distance = np.sqrt((X - x_average[0]) ** 2 + (Y - (x_average[1] * h))**2)
long_distance = np.sqrt((X - x_average[0]) ** 2 + (Y - (x_average[1] * h) + h)**2)

angle = np.arcsin(flat_distance/long_distance)

for i in x:
    plt.plot(i[0], i[1], 'x')
    
plt.pcolor(X, Y, -(X - x_average[0]) ** 2 - (Y - (x_average[1] * h))**2, cmap='turbo')
plt.show()
plt.pcolor(X, Y, angle)
plt.show()
```
