```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FuncFormatter
import scienceplots
plt.style.use('science')
plt.rcParams["figure.figsize"] = (4, 3)
plt.rcParams["figure.dpi"] = 150
```

```python
x = np.linspace(0, 2*np.pi, 400)
t_s = 10
t_dot = 1.5
t_dotsq = 3.2

agility_linear = t_s + t_dot * x
agility_bangbang = t_s + t_dotsq * np.sqrt(x)

print(x)
```

```python
def format_func(value, tick_number):
    N = int(np.round(value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi$"
    elif N == -1:
        return r"-$\pi$"
    else:
        return r"${0}\pi$".format(N)


plt.plot(x, agility_linear, 'k', linewidth=1.5, label="Linear Agility")
plt.plot([0, 0], [0, agility_linear[0]], 'k--', linewidth=0.75)
plt.plot([x[-1], x[-1]], [0, agility_linear[-1]], 'k--', linewidth=0.75)

plt.gca().xaxis.set_major_formatter(FuncFormatter(format_func))
plt.gca().xaxis.set_major_locator(MultipleLocator(base=np.pi))

plt.ylim(0)
plt.xlim(0)
plt.ylabel("Time [s]")
plt.xlabel("Slew Angle [rad]")
plt.legend()
plt.tight_layout()
plt.savefig("linear_agility.pdf")
plt.show()
```
