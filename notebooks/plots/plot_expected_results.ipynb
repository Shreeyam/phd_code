```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scienceplots 
plt.style.use(["science", "grid"])
```

```python
# Data points
import matplotlib.patches


x_values = np.array([1, 0.9, 0.9, 0.7])
y_values = np.array([0.33, 0.5, 0.7, 0.9])
labels = ['Conventional', '1-shot lookahead', 'Always-on lookahead', 'Reinforcement learning']
colors = ['blue', 'green', 'red', 'purple']
markers = ['o', 's', 'D', 'p']


# Plot
plt.figure(figsize=(5, 3.5), dpi=150)
ax=plt.gca()

# Region of improvement
patch = matplotlib.patches.Polygon([ [1, 0.33], [1, 1], [0.33, 1]], closed=True, edgecolor='none', facecolor='#c1c3ec', linestyle='--', label="Improvement region")
ax.add_patch(patch)
for i, label in enumerate(labels):
    scatter = ax.scatter(x_values[i], y_values[i], c=colors[i], label=label, marker=markers[i], s=40, zorder=5)
    scatter.set_clip_on(False)

# Omniscient point
ax.scatter(1, 1.0, c='orange', label='Omniscient', marker='*', s=100, zorder=5).set_clip_on(False)

# Labels and Title
ax.set_xlabel('Total Captured Images [-]')
ax.set_ylabel('Proportion of Cloud-Free Images [-]')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xticks([0.33, 1])
ax.set_xticklabels(['~0.33$X_0$', '$X_0$'])
ax.set_yticks([0.33, 1])
ax.set_yticklabels(['0.33', '1'])

plt.legend(ncol=3, loc='lower center', bbox_to_anchor=(0.5, 1.02), fontsize=7)
plt.tight_layout()
plt.show()
```
